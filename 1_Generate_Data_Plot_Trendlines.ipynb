```python
from glob import glob
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
%matplotlib inline
```

```python
csv_output_f = './data/reproduced/'
```

```python
if not os.path.exists(csv_output_f):
    os.makedirs(csv_output_f)
```

## Get files names

```python
all_files = glob('data/components/*')
community_files = glob(os.path.join('data/components', "?_community.csv"))
noncommunity_files = glob(os.path.join('data/components', "?_nonCommunity.csv"))
```

### Read community

```python
list_ids = []
community_df = pd.DataFrame()
for file_ in community_files:
    df = pd.read_csv(file_, index_col=0)
    community_df = pd.concat([community_df, df])
    for id_ in df.index.values:
        list_ids.append(id_)
    print(file_, str(len(df.index.values)))
```

```python
print(len(list_ids), len(set(list_ids)))
```

```python
community_df.head()
```

### Read non community

```python
list_ids = []
noncommunity_df = pd.DataFrame()
for file_ in noncommunity_files:
    df = pd.read_csv(file_, index_col=0)
    noncommunity_df = pd.concat([noncommunity_df, df])
    for id_ in df.index.values:
        list_ids.append(id_)
    print(file_, str(len(df.index.values)))
```

```python
noncommunity_df.head()
```

### Save files

```python
community_df.to_csv(csv_output_f+'all_community.csv')
noncommunity_df.to_csv(csv_output_f+'all_noncommunity.csv')
```

## Indexing and verifying missing data

```python
community_df.columns = community_df.columns.astype(int)
noncommunity_df.columns = noncommunity_df.columns.astype(int)
```

```python
community_df = community_df.reindex(
                    sorted(community_df.columns), axis=1)
noncommunity_df = noncommunity_df.reindex(
                    sorted(noncommunity_df.columns), axis=1)
```

### Save data without NaNs

```python
community_df.to_csv(csv_output_f+'no_nan_community.csv')
noncommunity_df.to_csv(csv_output_f+'no_nan_noncommunity.csv')
```

### Plotting

```python
x_com = community_df.columns.values
x_com = [x+1 for x in x_com]
y_com = community_df.mean().values

# calc the trendline
z_com = np.polyfit(x_com, y_com, 1)
p_com = np.poly1d(z_com)
# the line equation:
print("y_com=%.6fx_com+(%.6f)"%(z_com[0],z_com[1]))
```

```python
x_non = noncommunity_df.columns.values
x_non = [x+1 for x in x_non]

y_non = noncommunity_df.mean().values

# calc the trendline
z_non = np.polyfit(x_non, y_non, 1)
p_non = np.poly1d(z_non)
print("y_non=%.6fx_non+(%.6f)"%(z_non[0],z_non[1]))
```

```python
plt.figure(figsize=((16,8)))
plt.xlim(1, 84)
# Plot community means
plt.scatter(x_com, y_com, marker='*', c='b')
plt.plot(x_com,p_com(x_com),"b--")
# Plot non community means
plt.scatter(x_non, y_non, c='r', marker='x')
plt.plot(x_non,p_non(x_non),"r--")
plt.xlabel('Days in plan')
plt.ylabel('Mean PAL')
```

```python
community_df.mean().to_csv(csv_output_f+'mean_community.csv')
noncommunity_df.mean().to_csv(csv_output_f+'mean_noncommunity.csv')
```
